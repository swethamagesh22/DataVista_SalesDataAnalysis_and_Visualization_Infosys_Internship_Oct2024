```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure numeric_data is defined correctly
# Select only numeric columns from your dataset
numeric_data = vehicles.select_dtypes(include='number')  # Ensure 'vehicles' is a valid DataFrame

# Set up a 1x2 grid for pie chart and heatmap
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Pie chart for distribution of EV modes (unchanged)
mode_counts = vehicles['mode'].value_counts()
axes[0].pie(mode_counts, labels=mode_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
axes[0].set_title('Distribution of EV Modes')
axes[0].axis('equal')  # Ensures the pie is drawn as a circle

# Correlation heatmap for numeric columns
if not numeric_data.empty:
    correlation_matrix = numeric_data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', ax=axes[1])
    axes[1].set_title('Heatmap of Correlations')
else:
    print("No numeric data available for correlation.")

# Adjust layout to prevent overlap and display all plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up a 3x2 grid for all plots
fig, axes = plt.subplots(2, 3, figsize=(20, 12))

# Flatten axes array for easy indexing
axes = axes.ravel()

# 1. Line plot for mode count distribution
mode_counts = vehicles['mode'].value_counts()
axes[0].plot(mode_counts.index, mode_counts.values, marker='o', linestyle='-', color='b')
axes[0].set_title('Mode Count Distribution (Line Plot)')
axes[0].set_xlabel('Mode Type')
axes[0].set_ylabel('Count')
axes[0].grid(True)

# 2. Area plot of mode counts
axes[1].fill_between(range(len(mode_counts)), mode_counts, color="skyblue", alpha=0.4)
axes[1].plot(mode_counts, color="Slateblue", alpha=0.6, linewidth=2)
axes[1].set_title('Area Plot of Mode Counts')
axes[1].set_xlabel('Mode Type')
axes[1].set_ylabel('Count')

# 3. Scatter plot for mode vs. count
modes = vehicles['mode']
counts = vehicles['value']  # Replace with your actual numeric column
axes[2].scatter(modes, counts, color='g', alpha=0.6)
axes[2].set_title('Mode vs. Count (Scatter Plot)')
axes[2].set_xlabel('Mode Type')
axes[2].set_ylabel('Value')
axes[2].grid(True)

# 4. Box plot of counts by mode
sns.boxplot(ax=axes[3], x='mode', y='value', data=vehicles)
axes[3].set_title('Box Plot of Counts by Mode')
axes[3].set_xlabel('Mode Type')
axes[3].set_ylabel('Value')

# 5. Line plot for 'value' over 'year', grouped by 'mode'
sns.lineplot(ax=axes[4], data=vehicles, x="year", y="value", hue="mode", marker="o")
axes[4].set_title("Trend of EV Values Over Years by Mode")
axes[4].set_xlabel("Year")
axes[4].set_ylabel("Value")

# 6. Strip plot for 'value' by 'mode'
sns.stripplot(ax=axes[5], data=vehicles, x="mode", y="value", hue="mode", palette="Set2", jitter=True, dodge=True, legend=False)
axes[5].set_title("Value Distribution by Mode (Strip Plot)")
axes[5].set_xlabel("Mode Type")
axes[5].set_ylabel("Value")

# Adjust layout to prevent overlap and show the plots
plt.tight_layout()
plt.show()
```
